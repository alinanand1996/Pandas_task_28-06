Create a new column 'income_binary' where <=50K is 0 and >50K is 1.

```python
def inc_bin(value):
  if value == '<=50K':
    return(0)
  else:
    return(1)
data['income_binary']=data['income'].map(inc_bin)
data

```

Create a new column 'age_group' categorizing ages into 'Youth' (< 30), 'Adult' (30-60), and 'Senior' (60+).

```python
def catogarize(value):
  if value < 30:
    return('Youth')
  elif value >=30 and value <60:
    return('adult')
  elif value >=60:
    return('senior')
data['age_group']=data['age'].map(catogarize)
data
#data.iloc[data['age'].idxmax()]
```
